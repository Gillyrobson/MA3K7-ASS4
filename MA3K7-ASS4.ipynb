```python
def coin_flips(num_flips):
    seq_tot_25 = 0 
    for i in range(2**num_flips):
        flips = bin(i)[2:].zfill(num_flips)  
        total = 0
        for flip in flips:
            total += int(flip) + 1
            if total == 25:
                seq_tot_25 += 1
                break
    return seq_tot_25

num_flips = 24
num_seq = coin_flips(num_flips)+1

print("Number of sequences with total 25:", num_seq)
```

```python
def P(n):
    if n == 0:
        return 1  
    else:
        return 1 - 0.5 * P(n-1)


n = 25
result = P(n)
print("P({}) = {:.7f}".format(n, result))
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

def P(n):
    if n == 1:
        return 1  
    else:
        return 1 - 0.5 * P(n-1)


n_val1 = list(range(1, 26))  
p_val1 = [P(n) for n in n_val1]


sns.set_style("whitegrid")
plt.figure(figsize=(10, 6))
plt.plot(n_val1, p_val1, marker='o', linestyle='-', color='pink')
plt.title('Graph of P(n)')
plt.xlabel('n')
plt.ylabel('P(n)')
plt.xticks(n_val1)
plt.grid(True)
plt.show()
```
